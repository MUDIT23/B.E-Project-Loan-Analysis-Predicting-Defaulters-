2. Extracting Values

```python
# Extracting the number from the string value
# e.g '10+ years' is converted to 10

data1['emp_length'] = data1['emp_length'].str.split(' ').str[0]
#print(data1['emp_length'].value_counts())

# Replacing the values
# '10+' -> 10
# '<' -> 0

data1['emp_length'].loc[data1['emp_length'] == '10+'] = 10
data1['emp_length'].loc[data1['emp_length'] == '<'] = 0
data1['emp_length'].value_counts()
```

```python
# Extracting the number from the string value
# e.g ' 36 months' -> 36

#data1['term'].value_counts()
data1['term'] = data1['term'].str.split(' ').str[1]
data1['term'].value_counts()
```

```python
# extracting the month and year value for each date variable
# e.g '01-02-1985' (dd-mm-yyyy) -> month -> 2
#                                  year -> 1985


dv = ['first_credit','issue_date','credit_report','last_payment_date','next_payment_date']

for i in dv:
    data1[i] = pd.to_datetime(data1[i], format = '%d-%m-%Y')
    a = i+'_month'
    b = i+'_year'
    data1[a] = data1[i].apply(lambda date : date.month)
    data1[b] = data1[i].apply(lambda date : date.year)


# dropping the original date columns except for 'issue_date'

dv.remove('issue_date') 
data1.drop(dv, axis = 1, inplace = True)
```

Imporing Libraries TO Apply Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import auc, accuracy_score, precision_score, recall_score, f1_score, \
                            roc_curve,classification_report, confusion_matrix
```

Function To Calculate Precessioon, Recall, F1 - Score, Area Under Curve (AUC)

```python
def metric(mdl,xtest,ytest):
    l = []
    
    # AUC
    fpr, tpr, thresholds = roc_curve(ytest, mdl.predict(xtest))
    l.append(round(auc(fpr,tpr),2))
    #print(auc(fpr,tpr))
    
    # CA
    #print(round(accuracy_score(ytest,mdl.predict(xtest)),4))
    l.append(round(accuracy_score(ytest,mdl.predict(xtest)),2))
    
    # Confusion matrix
    cfm = confusion_matrix(ytest,mdl.predict(xtest))
    
    #print(cfm)
    
    # cfm[0,0] -> TN
    # cfm[0,1] -> FP
    # cfm[1,0] -> FN
    # cfm[1,1] -> TP
    

    # Precision class 0
    p0 = round(cfm[0,0]/(cfm[0,0]+cfm[1,0]),2)
    l.append(p0)
    
    # recall class 0
    r0 = round(cfm[0,0]/(cfm[0,0]+cfm[0,1]),2)
    l.append(r0)
    
    # F1 Score 0
    f1_0 = round((2*p0*r0)/(p0+r0),2)
    l.append(f1_0)
    
    
    l.append(round(auc(fpr,tpr),2))
    l.append(round(accuracy_score(ytest,mdl.predict(xtest)),2))
    
    # Precision class 1
    #print(round(precision_score(ytest,mdl.predict(xtest)),4))
    l.append(round(precision_score(ytest,mdl.predict(xtest)),2))
    
    # Recall class 1
    #print(round(recall_score(ytest,mdl.predict(xtest)),4))
    l.append(round(recall_score(ytest,mdl.predict(xtest)),2))
    
    # F1 Score class 1
    #print(round(f1_score(ytest,mdl.predict(xtest)),4))
    l.append(round(f1_score(ytest,mdl.predict(xtest)),2))

    '''
    if mdl == lr:
        a = 'Logistic'
    elif mdl == dt:
        a = 'Decision Tree'
    else:
        a = 'AdaBoost'

    if b == 1:
        tr = 'Normal Data'
    elif b == 2:
        tr = 'Smote Data'
    else:
        tr = 'Smote ENN Data'
    
    print('Classification Report for',a,'using',tr,'\n')
    
    print(classification_report(ytest,mdl.predict(xtest)))
    '''
    
    return l
```

Function For Applying Algorithms 

```python
def mod(mdl,xtrain,ytrain,xtest,ytest,tt):
    mdl.fit(xtrain,ytrain)
    
    if mdl == lr:
        a = 'Logistic Regression'
    elif mdl == dt:
        a = 'Decision Tree'
    else:
        a = 'AdaBoost'
    
    l1 = metric(mdl,xtest,ytest)
    
    for k,v in enumerate(l1):
        
        k1 = k
        if k > 4:
            b = 1
            k1 = k-5 
        else:
            b = 0
            k1 = k
        
        #print(k,v)
        if tt == 0:
            result.loc[(a,b),result.columns[k1]] = v
        elif tt == 1:
            result1.loc[(a,b),result.columns[k1]] = v
        elif tt == 2:
            result2.loc[(a,b),result.columns[k1]] = v
        elif tt == 3:
            result3.loc[(a,b),result.columns[k1]] = v
        else:
            result4.loc[(a,b),result.columns[k1]] = v
    
```

Passing Values To The MOD Function

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier(random_state = 10)
ada = AdaBoostClassifier(n_estimators = 100, random_state = 10)

for i in [lr,dt,ada]:

    # Unbalanced data
    mod(i,X_train,Y_train,X_test,Y_test,0)

    # smote train data
    mod(i,X_train_se,Y_train_se,X_test,Y_test,1)

    # Smote ENN train data
    mod(i,X_train_sme_enn,Y_train_sme_enn,X_test,Y_test,2)
    
    # Whole smote data
    mod(i,X_train_se_wh,Y_train_se_wh,X_test_se_wh,Y_test_se_wh,3)

    # Whole Smote ENN data
    mod(i,X_train_sme_enn_wh,Y_train_sme_enn_wh,X_test_sme_enn_wh,Y_test_sme_enn_wh,4)
```
